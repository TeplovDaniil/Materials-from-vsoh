```python
le = LabelEncoder()
df['y']=le.fit_transform(df['y'])
```

```python
X = df.drop(columns=["y", "id"])
y = df["y"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,stratify=y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
models = VotingClassifier(
    estimators=[
        ('svc', SVC(probability=True, gamma='auto')),
        ('rf', RandomForestClassifier(n_jobs=-1, n_estimators=25)),
        ('xgb', XGBClassifier(n_jobs=-1, n_estimators=250, max_depth=6)),
        ('knn', KNeighborsClassifier(n_jobs=-1, n_neighbors=10)),
        ('catboost', CatBoostClassifier(verbose=0, iterations=200,depth=8,task_type='CPU')),
        # ('lgbm', LGBMClassifier(n_jobs=-1)),
    ],
    voting='hard',
    n_jobs=-1,
    verbose=True
)
models.fit(X, y)

```
